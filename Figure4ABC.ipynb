```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import scipy.stats as stats
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
from scipy.stats import kruskal
from scipy.stats import linregress
data = pd.read_excel(r"C:\Users\ç„¶\Desktop\sea-ad_cohort_donor_metadata_072524.xlsx")
```

```python
dementia_age = data.iloc[:, 19]
neuron_ratios = data.iloc[:, 69]
mmse_scores = pd.to_numeric(dementia_age, errors='coerce')
neuron_ratios = pd.to_numeric(neuron_ratios, errors='coerce')
data = pd.DataFrame({'dementia_age': dementia_age, 'Neuron': neuron_ratios}).dropna()
print(data)
```

```python
dementia_age = data['dementia_age']
neuron_ratios = data['Neuron']
slope, intercept, r_value, p_value, std_err = linregress(dementia_age, neuron_ratios)
fit_line = slope * dementia_age + intercept

plt.figure(figsize=(8, 6))
plt.scatter(dementia_age, neuron_ratios, color='blue', label='Data Points')
plt.plot(dementia_age, fit_line, color='red', label=f'Fit Line: y={slope:.2f}x+{intercept:.2f}')
plt.title('dementia_age vs IL2/3 Ratios', fontsize=14)
plt.xlabel('dementia_age', fontsize=12)
plt.ylabel('Neuron Ratios', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R-squared: {r_value**2:.4f}")
print(f"P-value: {p_value:.4e}")
plt.show()
```
